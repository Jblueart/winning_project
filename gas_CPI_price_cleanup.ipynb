## Create a gas price Dataframe from weekly to monthly
### Gas price will show with monthly

```python
# Create a new column to specify the month from the dates
gas_df["Month"]=gas_df["Date"].dt.to_period("M")
```

```python
gas_df.head(10)
```

```python
# Create update dataframe - transform the weekly gas price to monthly by getting the mean from weekly gas price
gas_price_mean=gas_df.groupby(["Month"]).mean()
gas_price_mean.head()
```

```python
# Make Date index column
gas_price_mean.index.names = ['Date']
gas_price_mean.head()
```

```python
# Export the dataframe to csv file
gas_price_mean.to_csv("gas_price.csv")
```
